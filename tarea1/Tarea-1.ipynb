```python
import pandas as pd
from sklearn import preprocessing
import numpy as np
```

```python
dataset = pd.read_csv("daily-min-temperatures.csv")
```

```python
min = dataset['Temp'].min()
max = dataset['Temp'].max()
def norm(x):
    return (x-min)/(max-min)
```

```python
dataset['Temp'].std(ddof=1)
```

```python
mean = dataset['Temp'].mean()
std = dataset['Temp'].std()
def standard(x):
    return (x-mean)/std
```

```python
dataset['Temp_Norm_Manual'] = dataset['Temp'].apply(norm)
dataset ['Temp_Estandar_Manual'] = dataset['Temp'].apply(standard)
```

```python
# Es necesario extraer los valores del columna del dataset Temp, y luego convertirlo todos en row, ya que si no no los puede trabajar como dentro de una columna
dataset['Temp_Norm_SK'] = preprocessing.MinMaxScaler().fit_transform(dataset['Temp'].values.reshape((-1, 1)))
dataset['Temp_Estandar_SK'] = preprocessing. StandardScaler().fit_transform(dataset['Temp'].values.reshape((-1, 1)))
```
