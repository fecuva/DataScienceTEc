# Introduccion 
Esta base de datos fue tomada de una empresa minera en Canada, especialmente plantas mineras. Y proviene de una de las partes mas importantes del proceso, la planta de flotación, en el que se separan los sedimentos y se obtiene el mineral puro. 

La flotación por espuma es un proceso para separar selectivamente materiales hidrófobos de hidrófilos. Esto se utiliza en las industrias de procesamiento de minerales, reciclaje de papel y tratamiento de aguas residuales. Históricamente, esto se usó por primera vez en la industria minera, donde fue una de las grandes tecnologías habilitadoras del siglo XX. Se ha descrito como "la operación más importante utilizada para la recuperación y mejora de minerales de sulfuro". El desarrollo de la flotación de espuma ha mejorado la recuperación de minerales valiosos, como los minerales que contienen cobre y plomo. Junto con la minería mecanizada, ha permitido la recuperación económica de metales valiosos de  porcentajes muy bajos que los proceso's utilizados anteriormente.

## Descripcion del proceso 

Antes de que la flotación de poder ejecutar el proceso, el mineral a tratar se reduce a partículas finas triturando y moliendo (un proceso conocido como conminución) para que los diversos minerales existente se separen como granos físicamente separados. Este proceso se conoce como liberación. Los tamaños de partícula son típicamente menores de 0.1 mm (100 µm), pero a veces se requieren tamaños menores de 7–10 µm. Hay una tendencia a que el tamaño de liberación de los minerales disminuya con el tiempo a medida que los cuerpos de minerales con granos gruesos -que pueden separarse en tamaños más grandes- se agotan y reemplazan por cuerpos minerales que anteriormente se consideraban demasiado difíciles.

En la industria minera, las plantas donde se realiza la flotación para concentrar mineral se conocen generalmente como concentradores o molinos.

Para la flotación, el mineral molido se mezcla con agua para formar una suspensión y el mineral deseado se vuelve hidrofóbico mediante la adición de un agente tensioactivo o químico colector (aunque algunas superficies minerales son hidrofóbicas naturalmente, que requieren poca o ninguna adición de colector ). La sustancia química particular depende de la naturaleza del mineral a recuperar y, tal vez, de la naturaleza de los que no se desean. 

Como ejemplo, se puede agregar etil xantato de sodio como un colector en la flotación selectiva de galena (sulfuro de plomo) para separarlo de la esfalerita (sulfuro de zinc). Esta suspensión (más propiamente llamada pulpa) de partículas hidrofóbicas y partículas hidrofílicas se introduce luego en tanques conocidos como células de flotación que se airean para producir burbujas. Las partículas hidrofóbicas se adhieren a las burbujas de aire, que se elevan a la superficie, formando una espuma. La espuma se elimina de la célula, produciendo un concentrado del mineral objetivo.

Se pueden introducir agentes espumantes en la pulpa para promover la formación de una espuma estable en la parte superior de la celda de flotación.

Los minerales que no flotan en la espuma se denominan relaves de flotación o colas de flotación. Estos relaves también pueden someterse a otras etapas de flotación para recuperar las partículas valiosas que no flotaron la primera vez. Esto se conoce como barrido. Los relaves finales después del barrido normalmente se bombean para su eliminación como relleno de mina o en instalaciones de eliminación de relaves para el almacenamiento a largo plazo.

La eficiencia de la flotación de espuma está determinada por una serie de probabilidades: las de contacto entre partículas y burbujas, la unión entre partículas y burbujas, el transporte entre la pulpa y la espuma, y la recolección de espuma en el lavado del producto. En una celda convencional agitada mecánicamente, la fracción vacía (es decir, el volumen ocupado por las burbujas de aire) es baja (5 a 10 por ciento) y el tamaño de la burbuja suele ser mayor de 1 mm. Esto da como resultado un área interferencia relativamente baja y una baja probabilidad de contacto entre partículas y burbujas. En consecuencia, se requieren varias celdas en serie para aumentar el tiempo de residencia de la partícula, aumentando así la probabilidad de contacto de la partícula-burbuja. 

La flotación se lleva a cabo normalmente en varias etapas para maximizar la recuperación del mineral o minerales objetivo y la concentración de esos minerales en el concentrado, mientras se minimiza el aporte de energía.

### Etapas de la flotacion 

- Desbaste

    La primera etapa se llama desbaste, que produce un concentrado más áspero. El objetivo es eliminar la cantidad máxima de mineral valioso en un tamaño de partícula tan grueso como sea práctico. Cuanto más fino se muele un mineral, mayor será la energía necesaria, por lo que tiene sentido moler solo aquellas partículas que necesitan molienda fina.
- Limpieza

    El concentrado más áspero normalmente se somete a otras etapas de flotación para rechazar más de los minerales indeseables que también formaron parte de la espuma, en un proceso conocido como limpieza. El producto de limpieza se conoce como el concentrado limpio o el concentrado final. El objetivo de la limpieza es producir una concentration del mineral objetivo tan alto como sea posible.
- Barrido

    El paso de flotación más áspero a menudo es seguido por un paso de flotación del carroñero que se aplica a los relaves más ásperos. El objetivo es recuperar cualquiera de los minerales objetivo que no se recuperaron durante la etapa inicial de desbaste.

```python
from sklearn.decomposition import PCA
componentes = 10
pca = PCA(n_components = componentes)
print(pca)
```

```python
explained_variance = pca.explained_variance_ratio_
d = pd.DataFrame({'factors':['factor_'+str(n+1) for n in range(componentes)],'variance':explained_variance})
d.plot.line(x = 'factors')
```

A partir de lo anterior se puede observar que los factores no son capaces de explicar más del 24% de la varianza de los componentes que se asocian para cada factor no obstante si se realiza para dos factores que son los que más logran agrupar capturar la varianza del dataset entonces se puede ver que a su vez los features no explican en gran medida a cada factor. 

```python
componentes = 2
pca = PCA(n_components = componentes)
print(pca)
pca.fit_transform(X_train)
pca.transform(X_test)
dataset = {'feature':cols[1:-1]}
for n in range(componentes): dataset['factor_'+str(n+1)] = pca.components_[n]
d = pd.DataFrame(dataset)
d.dtypes
d[(d['factor_2']<-0.30) | (d['factor_2']>0.30)|(d['factor_1']>0.30)|(d['factor_1']<-0.30)]
```

## Red Neuronal.

A pesar que obtuvimos un modelo que generaliza eficientemente el set de datos, se desarrollo una red neuronal MLP, con el fin de revisar si este modelo tieme mejores o iguales resultados al de Kesimo vecino mas cercano.


### Division del set de datos para la red neuronal

Se ejecuta la division del set entre training y test, se usa el definido por la libreria pues se cuenta con una cantidad muy Grande de datos. 

```python
import keras
from keras.utils import np_utils

X =  X_train.values

y = Y_train.values

print("Shape of X:", X.shape)
print("\nShape of y:", y.shape)
print("\nFirst 10 rows of X")
print(X[:10])
print("\nFirst 10 rows of y")
print(y[:10])
```

```python
import warnings 
#warnings.filterwarnings("ignore")
import numpy as np
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation
from keras.optimizers import SGD
from keras.optimizers import Adam
from keras.utils import np_utils
from keras import optimizers

# Building the model

model = Sequential()      #Utilizamos un modelo secuencial, segun el modelo visto en clase

model.add(Dense(1024, input_dim=22))
model.add(Activation('sigmoid'))

model.add(Dense(512))
model.add(Activation('tanh'))# intentos fallidos de capas, la red no convergio

model.add(Dense(256))
model.add(Activation('tanh'))#intentos fallidos de capas, la red no convergio

model.add(Dense(128))
model.add(Activation('sigmoid'))#intentos fallidos de capas, la red no convergio

model.add(Dense(64))
model.add(Activation('sigmoid'))#intentos fallidos de capas, la red no convergio

model.add(Dense(1))
model.add(Activation('sigmoid'))

# Compiling the model

sgd = optimizers.SGD(lr=0.00001, decay=0.000001, momentum=0, nesterov=False)
# adam = optimizers.Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
model.compile(loss = 'mean_squared_error', optimizer= sgd, metrics=['accuracy'])


model.summary()
```

Se utiliza un epoch de 500 y un batch size de 1000, debido a la cantidad de datos que se Tienen. 

```python
model.fit(X, y, epochs=500, batch_size=1000, verbose=1)
```

```python
score = model.evaluate(X, y)
print("\n Training Accuracy:", score[1])
```

Se obtiene un  trainig score demasiado bajo. Esto se debe a que el problema es Altamente lineal y las redes neuronales son muy fuertes en problemas no lineales.
